### Part 1 (1 point)

Load the Helsinki temperature data from the file `data/helsinki-vantaa.csv`.

- Read the data into a variable called `data` using pandas
- Parse the dates from the column `'DATE'` and set the dates as the index in the DataFrame 

```python
# YOUR CODE HERE
# import pandas and data
import pandas as pd
data = pd.read_csv("data/helsinki-vantaa.csv", parse_dates = ["DATE"], index_col = "DATE")
```

### Part 2 (2 points)

Select data for the 30-year period (January 1988 to December 2018).

- Store the selection in a new variable `selection`

```python
# YOUR CODE HERE
# select data from specific time frame
selection = data.loc[(data.index >= "1988-01-01") & (data.index < "2019-01-01")]
```

```python
# Check that the data was read in correctly:
selection.head()
```

```python
# Check how many rows of data you selected:
print(f'Number of rows: {len(selection)}')
```

```python
%matplotlib inline

# YOUR CODE HERE
# import matplotlib
import matplotlib.pyplot as plt
# create figure
selection.plot(kind = "line", y = "TEMP_C", figsize = (14,6), 
               # adjust black line with circle marker
               style = "k-", marker = "o",
               # add title and axis labels
               title = "Helsinki-Vantaa Airport", xlabel = "Time", ylabel = "Temperature (Celsius)")
# add grid
plt.grid()

# Set output file name
outputfp = None
outputfp = "temp_line_plot.png"

# Save plot as image
# YOUR CODE HERE
outputfp = plt.savefig("temp_line_plot.png")
```
